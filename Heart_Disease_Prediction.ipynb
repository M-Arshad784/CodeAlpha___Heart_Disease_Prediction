## Check Outlier

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.xticks(rotation=60)
plt.show()
```

```python
def return_bounds(df, feature):
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return lower_bound, upper_bound
```

```python
!pip install feature_engine
```

```python
from feature_engine.outliers import ArbitraryOutlierCapper
```

```python
for feature in num_feature:
    lower_bound, upper_bound = return_bounds(df, feature)
    arb = ArbitraryOutlierCapper(min_capping_dict = {feature: lower_bound}, max_capping_dict = {feature: upper_bound})
    df = arb.fit_transform(df)
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(df)
plt.xticks(rotation=60)
plt.show()
```

## LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['Heart Disease'] = encoder.fit_transform(df['Heart Disease'])
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop('Heart Disease', axis=1),
                                                    df['Heart Disease'], test_size=0.2,
                                                    random_state=41)
```

## Scalling

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = ['Age', 'BP', 'Cholesterol', 'Max HR', 'ST depression']
scaler = StandardScaler()
```

```python
x_train[numeric_cols] = scaler.fit_transform(x_train[numeric_cols])
x_test[numeric_cols] = scaler.transform(x_test[numeric_cols])
```

## Model Traning

#### Install XGBoost

```python
!pip install xgboost
```

```python
import xgboost as xgb
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
```

```python
lr = LogisticRegression()
svm = SVC(kernel = 'poly', degree=2)
rf = RandomForestClassifier(n_estimators=500, random_state=42)
xg = XGBClassifier(n_estimators=100, max_depth=4, learning_rate=0.3, objective='binary:logistic', alpha=0.1)
```

```python
lr.fit(x_train, y_train)
svm.fit(x_train, y_train)
rf.fit(x_train, y_train)
xg.fit(x_train, y_train)
```

```python
lr_pred  = lr.predict(x_test)
svm_pred = svm.predict(x_test)
rf_pred  = rf.predict(x_test)
xg_pred  = xg.predict(x_test)
```

## Cross Val Score

```python
from sklearn.model_selection import cross_val_score
lr_score = np.mean(cross_val_score(lr, x_train, y_train, scoring='accuracy', cv=10))
print(f"Logistic Regression Cross Val Score : {lr_score}\n")
svm_score = np.mean(cross_val_score(svm, x_train, y_train, scoring='accuracy', cv=10))
print(f"SVM Cross Val Score : {svm_score}\n")
rf_score = np.mean(cross_val_score(rf, x_train, y_train, scoring='accuracy', cv=10))
print(f"Random Forest Cross Val Score : {rf_score}\n")
xg_score = np.mean(cross_val_score(xg, x_train, y_train, scoring='accuracy', cv=10))
print(f"XGBoost Cross Val Score : {xg_score}\n")
```

## Accuracy Score

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print(f"Logistic Regression Accuracy Score : {accuracy_score(y_test, lr_pred)}")
print(f"SVM Accuracy Score : {accuracy_score(y_test, svm_pred)}")
print(f"Random Forest Accuracy Score : {accuracy_score(y_test, rf_pred)}")
print(f"XGBoost Accuracy Score : {accuracy_score(y_test, xg_pred)}")
```
